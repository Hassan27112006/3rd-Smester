# STACK IMPLEMENTSTION

```python
class stack:
    def __init__(self,total_value):
        self.items=[]
        print("stack is ",self.items)
        self.limit=total_value  
        print("limit is ",self.limit)
    def push(self):
        if  self.limit==len(self.items):
            print("stack is full")
        else:
            for i in range(self.limit-len(self.items)):
                e=int(input("enter the element to be pushed"))
                self.items.append(e)
                print("after pushing the element stack is ",self.items)
    def pop(self):
        if len(self.items)==0:
            print("stack is empty")
        else:
            g=self.items.pop()
            print("popped element is ",g)
            print("after popping the element stack is ",self.items)
    def peek(self):
        if len(self.items)==0:
            print("stack is empty")
        else:
            print("top element is ",self.items[-1])
    def is_empty(self):
        if len(self.items)==0:
            print("stack is empty")
        else:
            print("stack is not empty")
    def is_full(self):
        if self.limit==len(self.items):
            print("stack is full")
        else:
            print("stack is not full")
    def sort(self):
        self.items.sort()
        print("stack in ascending order is ",self.items)
        self.items.sort(reverse=True)
        print("stack in descending order is ",self.items)

        
s=stack(5)








```

```python
while True:
    print("1.push\n2.pop\n3.peek\n4.is_empty\n5.is_full\n6.sort\n7.exit")
    choice=int(input("enter your choice"))
    if choice==1:
        s.push()
    elif choice==2:
        s.pop()
    elif choice==3:
        s.peek()
    elif choice==4:
        s.is_empty()
    elif choice==5:
        s.is_full()
    elif choice==6:
        s.sort()
    elif choice==7:
        print("system is exiting")
        break
    else:
        print("invalid choice")
        print("choose option from 1 to 7")
```
