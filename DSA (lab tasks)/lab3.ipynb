```python
def bubble_sort_desc(arr):
    n = len(arr)
    for i in range(n):
        for j in range(n-i-1):
            if arr[j] < arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
arr=[5,3,7,2,1]
bubble_sort_desc(arr)
```

```python
def bubble_sort_phases(arr2):
    n = len(arr2)
    for i in range(n):
        for j in range(n-1):
            if arr2[j] < arr2[j+1]:
                arr2[j], arr2[j+1] = arr2[j+1], arr2[j]
        print(f"Pass {i+1}: {arr2}")
    return arr2
arr2=[3,4,5,6,3,22,32]
bubble_sort_phases(arr2)
```

```python
import random

def bubble_sort_stats(arr):
    n = len(arr)
    comparisons = 0
    swaps = 0
    arr_copy = arr.copy()
    for i in range(n):
        for j in range(n-i-1):
            comparisons += 1
            if arr_copy[j] < arr_copy[j+1]:
                arr_copy[j], arr_copy[j+1] = arr_copy[j+1], arr_copy[j]
                swaps += 1
    return comparisons, swaps


sorted_arr = sorted(arr, reverse=True)
comp_sorted, swap_sorted = bubble_sort_stats(sorted_arr)


reverse_sorted_arr = sorted(arr)
comp_reverse, swap_reverse = bubble_sort_stats(reverse_sorted_arr)


random_arr = arr.copy()
random.shuffle(random_arr)
comp_random, swap_random = bubble_sort_stats(random_arr)

print("Sorted array:", sorted_arr)
print(f"Comparisons: {comp_sorted}, Swaps: {swap_sorted}\n")

print("Reverse-sorted array:", reverse_sorted_arr)
print(f"Comparisons: {comp_reverse}, Swaps: {swap_reverse}\n")

print("Random array:", random_arr)
print(f"Comparisons: {comp_random}, Swaps: {swap_random}")
```

```python
def bubble_sort_stats(arr3):
    n = len(arr3)
    comparisons = 0
    swaps = 0
    for i in range(n):
        for j in range(n-i-1):
            comparisons += 1
            if arr3[j] < arr3[j+1]:
                arr3[j], arr3[j+1] = arr3[j+1], arr3[j]
                swaps += 1
        print(f"Pass {i+1}: {arr3}")
    
    print(f"Total comparisons: {comparisons}")
    print(f"Total swaps: {swaps}")
    return arr3

arr3=[11,22,33,44,55,66,77,88]
bubble_sort_stats(arr3)
```

```python
def bubble_sort_optimized(arr5):
    n = len(arr5)
    arr_copy = arr5.copy()
    for i in range(n):
        swapped = False
        for j in range(n-i-1):
            if arr_copy[j] < arr_copy[j+1]:
                arr_copy[j], arr_copy[j+1] = arr_copy[j+1], arr_copy[j]
                swapped = True
        if not swapped:
            break
    return arr_copy


arr5=[21,34,45,56,67,78,89,90]
bubble_sort_optimized(arr5)
```
