```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" → ")
            current = current.next
        print("None")

    def reverse(self):
        prev = None
        current = self.head
        self.tail = self.head  

        while current:
            next_node = current.next   
            current.next = prev        
            prev = current           
            current = next_node       

        self.head = prev 


ll = LinkedList()
ll.append(10)
ll.append(20)
ll.append(30)
ll.append(40)

print("Original Linked List:")
ll.display()

ll.reverse()
print("Reversed Linked List:")
ll.display()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    def insertAtEnd(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
    def display(self):
        temp = self.head
        if not temp:
            print("List is empty")
            return
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

    def swapNodes(self, x, y):
        if x == y:
            print("Both nodes have same value — no swap needed.")
            return
        prevX = None
        currX = self.head
        while currX and currX.data != x:
            prevX = currX
            currX = currX.next

        prevY = None
        currY = self.head
        while currY and currY.data != y:
            prevY = currY
            currY = currY.next

        if not currX or not currY:
            print("One or both values not found — cannot swap.")
            return

        if prevX:
            prevX.next = currY
        else:
            self.head = currY

        if prevY:
            prevY.next = currX
        else:
            self.head = currX
        temp = currX.next
        currX.next = currY.next
        currY.next = temp

        print(f"Nodes with values {x} and {y} swapped successfully.")



ll = LinkedList()
ll.insertAtEnd(10)
ll.insertAtEnd(20)
ll.insertAtEnd(30)
ll.insertAtEnd(40)
ll.insertAtEnd(50)

print("Original Linked List:")
ll.display()

ll.swapNodes(20, 40)

print("\nLinked List after swapping nodes 20 and 40:")
ll.display()
```
