```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    # Hash Function → Based on number of letters
    def hash_function(self, word):
        return len(word) % self.size

    # Insert word using Linear Probing
    def insert(self, word):
        index = self.hash_function(word)

        # If slot is taken → linear probe
        start_index = index
        while self.table[index] is not None:
            index = (index + 1) % self.size

            # If we came back to starting position → table full
            if index == start_index:
                print("Hash Table is Full! Cannot insert:", word)
                return

        self.table[index] = word

    # Display hash table
    def display(self):
        for i, word in enumerate(self.table):
            print(f"Index {i}: {word}")
ht = HashTable(7)

ht.insert("cat")
ht.insert("apple")
ht.insert("ball")
ht.insert("dog")
ht.insert("sun")
ht.insert("moon")
ht.display()
```
