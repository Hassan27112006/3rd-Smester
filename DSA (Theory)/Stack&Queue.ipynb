```python
class Queue:
    def __init__(self, max_size):
        self.items = []
        self.max_size = max_size
        print("Queue created with max size of:", max_size)
    def enqueue(self,):
        if self.max_size==len(self.items):
            print("Queue is full")
        else:
            for i in range(self.max_size-len(self.items)):
                e=int(input("Enter element to be added: "))
                self.items.append(e)
                print("Element added:", e)
                print("Current Queue:", self.items)
    def dequeue(self):
        if len(self.items)==0:
            print("Queue is already empty")
        else:
            removed_item=self.items.pop(0)
            print("Element removed:", removed_item)
            print("Current Queue:", self.items)
    def front(self):
        if len(self.items)==0:
            print("No front element, Queue is empty")
        else:
            print("Front element is:", self.items[0])
    def rear(self):
        if len(self.items)==0:
            print("No rear element, Queue is empty")
        else:
            print("Rear element is:", self.items[-1])
    def is_full(self):
        if self.max_size==len(self.items):
            print("Queue is full")
        else:
            print("Queue is not full")
            print("Current Queue:", self.items)
    def is_empty(self):
        if len(self.items)==0:
            print("Queue is empty")
        else:
            print("Queue is not empty")
            print("Current Queue:", self.items)
    def sort(self):
        self.items.sort()
        print("Queue in aecending order :", self.items)
        print(f"Queue in decending order :{self.items[::-1]}")
        
obj1=Queue(3)
```

```python
class Browser:
    def __init__(self):
        self.history = []

    def visit(self, page):
        self.history.append(page)
        print(f"Visited: {page}")

    def back(self):
        if self.history:
            last_page = self.history.pop()
            print(f"Going back from: {last_page}")
        else:
            print("No pages left to go back.")

    def current_page(self):
        if self.history:
            print(f"Current page: {self.history[-1]}")
        else:
            print("No page open.")

    def show_history(self):
        if self.history:
            print("Browsing History:", " -> ".join(self.history))
        else:
            print("History is empty.")

# Example
browser = Browser()
browser.visit("google.com")
browser.visit("youtube.com")
browser.visit("github.com")
browser.current_page()
browser.back()
browser.current_page()
browser.show_history()
```

```python
class TextEditor:
    def __init__(self):
        self.changes = []

    def type(self, text):
        self.changes.append(text)
        print(f"Typed: {text}")

    def undo(self):
        if self.changes:
            removed = self.changes.pop()
            print(f"Undo: Removed '{removed}'")
        else:
            print("Nothing to undo.")

    def show_changes(self):
        if self.changes:
            print("All Changes:", " | ".join(self.changes))
        else:
            print("No changes made yet.")

# Example
editor = TextEditor()
editor.type("Hello")
editor.type("Hello World")
editor.show_changes()
editor.undo()
editor.show_changes()
```

```python
class Warehouse:
    def __init__(self, capacity):
        self.capacity = capacity
        self.boxes = []

    def add_box(self, box):
        if not self.is_full():
            self.boxes.append(box)
            print(f"Box {box} added.")
        else:
            print("Warehouse is full!")

    def ship_box(self):
        if not self.is_empty():
            shipped = self.boxes.pop()
            print(f"Box {shipped} shipped.")
        else:
            print("Warehouse is empty!")

    def is_full(self):
        return len(self.boxes) >= self.capacity

    def is_empty(self):
        return len(self.boxes) == 0

    def size(self):
        return len(self.boxes)

# Example
wh = Warehouse(3)
wh.add_box("Box1")
wh.add_box("Box2")
wh.add_box("Box3")
print("Is Full?", wh.is_full())
wh.ship_box()
print("Is Empty?", wh.is_empty())
print("Current Size:", wh.size())
```

```python
# Balanced Parentheses Checker
def is_balanced(expression):
    stack = []
    opening = "({["
    closing = ")}]"
    mapping = {')': '(', '}': '{', ']': '['}

    for char in expression:
        if char in opening:
            stack.append(char)  # push opening
        elif char in closing:
            if not stack or stack[-1] != mapping[char]:
                return False
            stack.pop()  # pop matching opening

    return len(stack) == 0


# Example usage
expr1 = "((a+b)*(c+d))"
expr2 = "((a+b)*(c+d)"

print(expr1, "→", "Balanced" if is_balanced(expr1) else "Not Balanced")
print(expr2, "→", "Balanced" if is_balanced(expr2) else "Not Balanced")
```

```python
# Reverse a String using Stack
def reverse_string(s):
    stack = []
    for char in s:
        stack.append(char)  # push
    
    reversed_str = ""
    while stack:
        reversed_str += stack.pop()  # pop
    
    return reversed_str


# Example usage
original = input("Enter a string: ")
reversed_str = reverse_string(original)

print("Original String:", original)
print("Reversed String:", reversed_str)
```

```python
from collections import deque

class TicketCounter:
    def __init__(self):
        self.queue = deque()

    def enqueue(self, customer):
        self.queue.append(customer)
        print(f"{customer} joined the queue.")

    def dequeue(self):
        if self.queue:
            served = self.queue.popleft()
            print(f"{served} has been served.")
        else:
            print("No customers to serve.")

    def peek(self):
        if self.queue:
            print("Next customer:", self.queue[0])
        else:
            print("No customers waiting.")

    def display(self):
        if self.queue:
            print("Waiting customers:", list(self.queue))
        else:
            print("No customers in queue.")


# Example usage
tc = TicketCounter()
tc.enqueue("Alice")
tc.enqueue("Bob")
tc.enqueue("Charlie")
tc.peek()
tc.dequeue()
tc.display()
```

```python
from collections import deque

class PrinterSpooler:
    def __init__(self):
        self.queue = deque()

    def add_job(self, job):
        self.queue.append(job)
        print(f"Job '{job}' added to the queue.")

    def print_job(self):
        if self.queue:
            current = self.queue.popleft()
            print(f"Printing job: {current}")
        else:
            print("No jobs to print.")

    def show_status(self):
        if self.queue:
            print("Currently printing:", self.queue[0])
            print("Jobs waiting:", list(self.queue)[1:])
        else:
            print("No jobs in the printer queue.")


# Example usage
ps = PrinterSpooler()
ps.add_job("Document1")
ps.add_job("Document2")
ps.add_job("Document3")
ps.show_status()
ps.print_job()
ps.show_status()
```

```python
from collections import deque

class HospitalQueue:
    def __init__(self):
        self.queue = deque()

    def add_patient(self, patient):
        self.queue.append(patient)
        print(f"Patient '{patient}' added to the queue.")

    def treat_patient(self):
        if self.queue:
            treated = self.queue.popleft()
            print(f"Currently treating: {treated}")
        else:
            print("No patients to treat.")

    def show_waiting(self):
        if self.queue:
            print("Patients waiting:", list(self.queue))
        else:
            print("No patients waiting.")


# Example usage
hq = HospitalQueue()
hq.add_patient("John")
hq.add_patient("Mary")
hq.add_patient("Alex")
hq.treat_patient()
hq.show_waiting()
```

```python
from collections import deque

class CallCenter:
    def __init__(self):
        self.queue = deque()

    def add_caller(self, name):
        self.queue.append(name)
        print(f"Caller '{name}' added to the queue.")

    def resolve_call(self):
        if self.queue:
            resolved = self.queue.popleft()
            print(f"Call resolved for: {resolved}")
        else:
            print("No calls to resolve.")

    def peek_next(self):
        if self.queue:
            print("Next caller:", self.queue[0])
        else:
            print("No callers waiting.")

    def display_callers(self):
        if self.queue:
            print("Callers waiting:", list(self.queue))
        else:
            print("No callers in queue.")


# Example usage
cc = CallCenter()
cc.add_caller("Sam")
cc.add_caller("Emma")
cc.add_caller("Liam")
cc.peek_next()
cc.resolve_call()
cc.display_callers()
```

```python
from collections import deque

class TaskScheduler:
    def __init__(self):
        self.queue = deque()

    def add_task(self, task):
        """Add a new task to the queue"""
        self.queue.append(task)
        print(f"Task '{task}' added to the scheduler.")
        self.show_count()

    def run_task(self):
        """Run (remove) the task at the front of the queue"""
        if self.queue:
            current = self.queue.popleft()
            print(f"Running task: {current}")
        else:
            print("No tasks to run.")
        self.show_count()

    def show_count(self):
        """Show number of tasks remaining"""
        print(f"Tasks remaining: {len(self.queue)}")

    def show_pending(self):
        """Display all pending tasks"""
        if self.queue:
            print("Pending tasks:", list(self.queue))
        else:
            print("No pending tasks.")


# Example usage
if __name__ == "__main__":
    scheduler = TaskScheduler()
    
    scheduler.add_task("Backup files")
    scheduler.add_task("Update_")
```
