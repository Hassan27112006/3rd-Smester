```python
import random

def merge_sort(arr):
    if len(arr) <= 1:
        return arr, 0, 0   
    
    mid = len(arr) // 2
    left, comp_left, merge_left = merge_sort(arr[:mid])
    right, comp_right, merge_right = merge_sort(arr[mid:])
    
    merged, comp_merge = merge(left, right)
    
    
    total_comparisons = comp_left + comp_right + comp_merge
    total_merges = merge_left + merge_right + 1
    
    return merged, total_comparisons, total_merges


def merge(left, right):
    result = []
    i = j = 0
    comparisons = 0
    
    while i < len(left) and j < len(right):
        comparisons += 1
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    result.extend(left[i:])
    result.extend(right[j:])
    return result, comparisons



def run_case(arr, name):
    sorted_arr, comparisons, merges = merge_sort(arr)
    print(f"\n{name} Input: {arr}")
    print(f"{name} Sorted: {sorted_arr}")
    print(f"Comparisons: {comparisons}, Merges: {merges}")


sorted_arr = [1,2,3,4,5,6,7,8,9]
run_case(sorted_arr, "Sorted Array")

reverse_sorted_arr = [9,8,7,6,5,4,3,2,1]
run_case(reverse_sorted_arr, "Reverse-Sorted Array")

random_arr = random.sample(range(1, 20), 10)
run_case(random_arr, "Random Array")
```
