```python
from collections import deque

def train_station_simulation(arrival_order):
    queue = deque(arrival_order)
    stack = []
    de_order = []
    expected = 1
    operations = []

    while queue or stack:
        if queue and queue[0] == expected:
            train = queue.popleft()
            operations.append(f" Train {train} departs.")
            de_order.append(train)
            expected += 1
        elif stack and stack[-1] == expected:
            train = stack.pop()
            operations.append(f"Train {train} leave the station.")
            de_order.append(train)
            expected += 1
        elif queue:
            train = queue.popleft()
            stack.append(train)
            operations.append(f"Train {train} moved to the main station.")
        else:

            break

    
    for op in operations:
        print(op)
    print("final order is:", de_order)
train_station_simulation(arrival_order=[4, 3, 2, 1, 5])
```
