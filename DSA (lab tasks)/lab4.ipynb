```python


def selection_sort(arr):
    comparisons = 0
    swaps = 0
    n = len(arr)
    a = arr.copy()
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            comparisons += 1
            if a[j] < a[min_idx]:
                min_idx = j
        if min_idx != i:
            a[i], a[min_idx] = a[min_idx], a[i]
            swaps += 1
    return comparisons, swaps


def bubble_sort(arr):
    comparisons = 0
    swaps = 0
    n = len(arr)
    a = arr[:]
    for i in range(n - 1):
        for j in range(n - 1 - i):
            comparisons += 1
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
                swaps += 1
    return comparisons, swaps


sorted_arr = [1, 2, 3, 4, 5]
reverse_arr = [5, 4, 3, 2, 1]
random_arr = [3, 1, 4, 5, 2]

test_cases = {
    "Sorted Array": sorted_arr,
    "Reverse Sorted Array": reverse_arr,
    "Random Array": random_arr
}

for case, arr in test_cases.items():
   
    s_comp, s_swaps = selection_sort(arr)
    b_comp, b_swaps = bubble_sort(arr)
    print(f"Selection Sort Comparisons: {s_comp}, Swaps: {s_swaps}")
    print(f"Bubble Sort Comparisons: {b_comp}, Swaps: {b_swaps}")

```

```python


def hybrid_sort(arr):
    comparisons = 0
    swaps = 0
    iterations = 0
    a = arr.copy()
    left, right = 0, len(a) - 1

    while left < right:
        min_idx = left
        max_idx = left

        for i in range(left, right + 1):
            comparisons += 2  
            if a[i] < a[min_idx]:
                min_idx = i
            if a[i] > a[max_idx]:
                max_idx = i

        
        if min_idx != left:
            a[left], a[min_idx] = a[min_idx], a[left]
            swaps += 1
            if max_idx == left:  
                max_idx = min_idx

       
        if max_idx != right:
            a[right], a[max_idx] = a[max_idx], a[right]
            swaps += 1

        left += 1
        right -= 1
        iterations += 1

    print("Sorted Array:", a)
    print("Iterations:", iterations)
    print("Comparisons:", comparisons)
    print("Swaps:", swaps)



arr = [364, 25, 12, 22, 11, 90, 45]
print("Original Array:", arr)
hybrid_sort(arr)
```
