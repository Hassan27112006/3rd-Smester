```python
from collections import deque

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:  # Ascending (1 = most critical)
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Example
patients_queue = deque([4, 2, 5, 1, 3])

# Step 1: Dequeue into list
patients_list = []
while patients_queue:
    patients_list.append(patients_queue.popleft())

# Step 2: Sort using insertion sort
sorted_patients = insertion_sort(patients_list)

# Step 3: Enqueue sorted patients back
for p in sorted_patients:
    patients_queue.append(p)

# Step 4: Display final queue
print("Final sorted patient queue (most critical first):")
while patients_queue:
    print(patients_queue.popleft(), end=" ")
```

```python
from collections import deque

def insertion_sort(arr, descending=False):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        if not descending:  # Ascending
            while j >= 0 and arr[j] > key:
                arr[j + 1] = arr[j]
                j -= 1
        else:  # Descending
            while j >= 0 and arr[j] < key:
                arr[j + 1] = arr[j]
                j -= 1
        arr[j + 1] = key
    return arr

# Example stack of roll numbers
scripts_stack = deque([105, 101, 110, 102, 108])

# Step 1: Pop into list
rolls = []
while scripts_stack:
    rolls.append(scripts_stack.pop())

# Step 2: Sort Ascending
asc_sorted = insertion_sort(rolls.copy(), descending=False)

# Step 3: Sort Descending
desc_sorted = insertion_sort(rolls.copy(), descending=True)

# Step 4: Push back Ascending order
asc_stack = deque()
for r in asc_sorted:
    asc_stack.append(r)

# Step 5: Push back Descending order
desc_stack = deque()
for r in desc_sorted:
    desc_stack.append(r)

# Display results
print("Ascending Order (pop stack):")
while asc_stack:
    print(asc_stack.pop(), end=" ")

print("\nDescending Order (pop stack):")
while desc_stack:
    print(desc_stack.pop(), end=" ")
```

```python
class Queue:
    def __init__(self, capacity=10):
        self.items = []
        self.capacity = capacity

    def enqueue(self, book_id):
        if not self.is_full():
            self.items.append(book_id)
        else:
            print("Queue is full!")

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        return None

    def peek(self):
        return self.items[0] if not self.is_empty() else None

    def is_empty(self):
        return len(self.items) == 0

    def is_full(self):
        return len(self.items) == self.capacity

    def display(self):
        print("Queue:", self.items)


class Stack:
    def __init__(self, capacity=10):
        self.items = []
        self.capacity = capacity

    def push(self, book_id):
        if not self.is_full():
            self.items.append(book_id)
        else:
            print("Stack is full!")

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        return None

    def peek(self):
        return self.items[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.items) == 0

    def is_full(self):
        return len(self.items) == self.capacity

    def display(self):
        print("Stack:", self.items)


def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr


# Example: Returned (Queue), Issued (Stack)
returned_books = Queue()
for book in [45, 12, 78]:
    returned_books.enqueue(book)

issued_books = Stack()
for book in [34, 90, 23]:
    issued_books.push(book)

# Step 1: Show initial state
print("Initial Returned Books:")
returned_books.display()
print("Initial Issued Books:")
issued_books.display()

# Step 2: Dequeue + Pop into lists
returned_list = []
while not returned_books.is_empty():
    returned_list.append(returned_books.dequeue())

issued_list = []
while not issued_books.is_empty():
    issued_list.append(issued_books.pop())

# Step 3: Merge lists
merged = returned_list + issued_list
print("\nMerged List:", merged)

# Step 4: Sort
sorted_books = insertion_sort(merged)
print("Sorted List:", sorted_books)

# Step 5: Reinsert
mid = len(sorted_books) // 2
for b in sorted_books[:mid]:
    returned_books.enqueue(b)
for b in sorted_books[mid:]:
    issued_books.push(b)

# Step 6: Show reinserted
print("\nReinserted Returned Books:")
returned_books.display()
print("Reinserted Issued Books:")
issued_books.display()

# Step 7: Display all again
print("\nDequeued Returned Books:")
while not returned_books.is_empty():
    print(returned_books.dequeue(), end=" ")

print("\nPopped Issued Books:")
while not issued_books.is_empty():
    print(issued_books.pop(), end=" ")
```
