```python
def bubble_sort(wait_times):
    n = len(wait_times)
    
    # Perform bubble sort
    for i in range(n - 1):  # Number of passes
        swapped = False
        print(f"Pass {i+1}: ", end="")
        
        for j in range(n - 1 - i):  # Last i elements are already sorted
            if wait_times[j] > wait_times[j + 1]:
                # Swap
                wait_times[j], wait_times[j + 1] = wait_times[j + 1], wait_times[j]
                swapped = True
        
        # Print the list after each pass
        print(wait_times)
        
        # If no swap happened â†’ already sorted
        if not swapped:
            break
    
    return wait_times


# Example test
wait_times = [45, 10, 30, 5, 60]
print("Original list:", wait_times)
sorted_times = bubble_sort(wait_times)
print("Final sorted list:", sorted_times)
```

```python
def bubble_sort_prices(prices):
    n = len(prices)
    for i in range(n - 1):  # All passes (no optimization)
        print(f"Pass {i+1}:")
        for j in range(n - 1 - i):
            if prices[j] > prices[j + 1]:
                print(f"  Swapping {prices[j]} and {prices[j+1]}")
                prices[j], prices[j + 1] = prices[j + 1], prices[j]
                print(f"  Current list: {prices}")
        print(f"End of pass {i+1}: {prices}\n")

# Example test
prices = [19.99, 5.50, 12.00, 8.75, 25.00]
bubble_sort_prices(prices)
print("Final sorted prices:", prices)
```

```python
def bubble_sort_scores(scores):
    n = len(scores)
    for i in range(n - 1):
        swapped = False
        print(f"Pass {i+1}:")
        for j in range(n - 1 - i):
            if scores[j] < scores[j + 1]:  # Descending
                print(f"  Swapping {scores[j]} and {scores[j+1]}")
                scores[j], scores[j + 1] = scores[j + 1], scores[j]
                swapped = True
                print(f"  Current list: {scores}")
        print(f"End of pass {i+1}: {scores}\n")
        if not swapped:
            break

# Example test
scores = [85, 92, 78, 95, 88, 70]
bubble_sort_scores(scores)
print("Final sorted scores:", scores)
```

```python
from collections import deque

def bubble_sort_logs_stack(logs_stack):
    # Pop all logs into a list
    logs_list = []
    while logs_stack:
        logs_list.append(logs_stack.pop())
    
    # Bubble sort (ascending, no optimization)
    n = len(logs_list)
    for i in range(n - 1):
        print(f"Pass {i+1}:")
        for j in range(n - 1 - i):
            if logs_list[j] > logs_list[j + 1]:
                print(f"  Swapping {logs_list[j]} and {logs_list[j+1]}")
                logs_list[j], logs_list[j + 1] = logs_list[j + 1], logs_list[j]
                print(f"  Current list: {logs_list}")
        print(f"End of pass {i+1}: {logs_list}\n")
    
    # Push sorted logs into new stack
    sorted_stack = deque()
    for log in logs_list:
        sorted_stack.append(log)
    
    return sorted_stack

# Example test
logs_stack = deque([404, 500, 200, 503, 200])
sorted_stack = bubble_sort_logs_stack(logs_stack)

print("Final sorted stack (popping elements):")
while sorted_stack:
    print(sorted_stack.pop(), end=" ")
```

```python
from collections import deque

def bubble_sort_queue(tasks_queue):
    # Step 1: Dequeue all tasks into a list
    tasks_list = []
    while tasks_queue:
        tasks_list.append(tasks_queue.popleft())
    
    # Step 2: Bubble sort (ascending, with swapped optimization)
    n = len(tasks_list)
    for i in range(n - 1):
        swapped = False
        for j in range(n - 1 - i):
            if tasks_list[j] > tasks_list[j + 1]:  # Ascending
                tasks_list[j], tasks_list[j + 1] = tasks_list[j + 1], tasks_list[j]
                swapped = True
        print(f"Pass {i+1}: {tasks_list}")
        if not swapped:  # Early stop if already sorted
            break
    
    # Step 3: Enqueue sorted tasks back into the queue
    sorted_queue = deque()
    for task in tasks_list:
        sorted_queue.append(task)
    
    return sorted_queue

# Example test
tasks_queue = deque([3, 1, 4, 2, 5])
sorted_queue = bubble_sort_queue(tasks_queue)

print("Final sorted queue (dequeueing elements):")
while sorted_queue:
    print(sorted_queue.popleft(), end=" ")
```
